```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_csv('datasets/census.csv')
df.head()
```

```python
# group by statename and return the average of census2010pop
for group, frame in df.groupby('STNAME'):
  avg = np.average(frame['CENSUS2010POP'])
  print('Counties in state ' + group + 
				' have an average population of ' + str(avg))
```

## Function as a parameter

 We want to return three groups group 0 has state name begin with M 
 second group 1 has state name begin with Q and the remain group.


```python
# First we have to set index to a column that we want to group
df = df.set_index('STNAME')
df.head()
```

```python
# we define a function 
def set_batch_number(item):
  if item[0] < 'M':
    return 0
  elif item[0] < 'Q':
    return 1
  else: 
    return 2

# pass a function to group by
for group, frame in df.groupby(set_batch_number):
  print('There are ' + 
				str(len(frame)) + ' records in group '
				+ str(group) + ' for processing')
```

## Group by with multi index

```python
df = pd.read_csv('datasets/listings.csv')
```

```python
# we will work with two columns cancellation_policy and review_scores_value
# we will set index to these column as a multi index
df = df.set_index(['cancellation_policy', 'review_scores_value'])
for group, frame in df.groupby(level=(0,1)):
  print(group)
```

```python
# now we want to return these column but seaparate it in 2 part one is 10 in review score and other is not
def grouping_fun(item):
  if item[1] == 10.0:
    return (item[0], '10.0')
  else:
    return (item[0], 'not 10.0')

for group, frame in df.groupby(by=grouping_fun):
  print(group)
```

## Aggregation

```python
df = df.reset_index()
# this will not ignore the nan value
df.groupby('cancellation_policy').agg({'review_scores_value': np.average})
```

```python
df.groupby('cancellation_policy').agg({'review_scores_value': 'mean'})
```

```python
df.groupby('cancellation_policy').agg({'review_scores_value': ['mean', 'std'], 'reviews_per_month': ['mean']})
```

# Transformation

```python
# This method will broadcasts the function you supply over the grouped DataFrame
cols = ['cancellation_policy', 'review_scores_value']
# transform it and store it in new data frame
transform_df = df[cols].groupby('cancellation_policy').transform('mean')
transform_df.head()
```

```python
# rename the column to be mean_scores_value and 
transform_df.rename({'review_scores_value': 'mean_scores_value'}, axis='columns', inplace=True)
transform_df.head(2)
```

```python
# merge it into the dataframe
df = df.merge(transform_df, left_index=True, right_index=True)
```

```python
df.loc[:, [ 'cancellation_policy', 'mean_scores_value', 'review_scores_value' ]].head()
```

```python
# We will create new column for the differences between review scores and mean scores
df['mean_diff'] = np.absolute(df['review_scores_value'] - df['mean_scores_value'])
```

```python
df.loc[:, [ 'cancellation_policy', 'mean_diff' ]].head()
```

# Filtering

```python
df.groupby('cancellation_policy').filter(lambda x : np.nanmean(x['review_scores_value']) > 9.2).head()
```
